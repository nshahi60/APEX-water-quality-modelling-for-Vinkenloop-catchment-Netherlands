```python
import os
import pandas as pd

file_path = "C:/Users/shahi/Downloads/APEX/Weather.txt"

data = pd.read_csv(file_path, delimiter=",", skiprows=13, skipinitialspace="True")
data=data.drop(labels="# STN", axis=1)
data
```

```python
data["RH"]=data["RH"].replace(-1, 0)
data[["FG_m/s", "TN_degC", "TX_degC", "RH_mm"]]=data[["FG", "TN", "TX", "RH"]]*0.1
data["Q_MJ/m2"]=data["Q"]*0.01
```

```python
weather=data[["YYYYMMDD","RH_mm", "TN_degC", "TX_degC", "Q_MJ/m2", "FG_m/s", "UG"]]
weather.to_csv("C:/Users/shahi/Downloads/APEX/weather.wth", sep=" ", index=False, header=False)
```
